```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
import warnings
```

```python
df_games = pd.read_csv('Datasets\games_limpios.csv')
df_reviews = pd.read_csv('Datasets\\reviews_limpios.csv')
df_items = pd.read_csv('Datasets\items_limpios.csv')
```

```python
print(f"Hay {df_games.shape[0]} filas y {df_games.shape[1]} columnas") 
```

```python
df_games.isnull().sum()
```

```python
df_games.describe(include='all') 
```

```python
# Histograma de la distribución de precios
plt.figure(figsize=(10, 6))
sns.histplot(df_games['price'], bins=20, kde=True, color='skyblue')
plt.title('Distribución de Precios')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Gráfico de barras para contar la cantidad de juegos por género
plt.figure(figsize=(10, 6))
sns.countplot(data=df_games, x='genres', hue='genres', palette='Set2', legend=False)
plt.title('Cantidad de Juegos por Género')
plt.xlabel('Género')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.show()
```

```python
# Boxplot para visualizar la distribución de precios por género
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_games, x='genres', y='price', palette='Set2')
plt.title('Distribución de Precios por Género')
plt.xlabel('Género')
plt.ylabel('Precio')
plt.xticks(rotation=45)
plt.show()
```

```python
# Ignorar las advertencias específicas
warnings.filterwarnings("ignore", message="Passing `palette` without assigning `hue` is deprecated*")
warnings.filterwarnings("ignore", message="When grouping with a length-1 list-like,*")
# Gráfico de barras para contar la cantidad de juegos con acceso temprano vs no acceso temprano
plt.figure(figsize=(8, 5))
sns.countplot(data=df_games, x='early_access', hue='early_access', palette='Set1', legend=False)
plt.title('Cantidad de Juegos con Acceso Temprano')
plt.xlabel('Acceso Temprano')
plt.ylabel('Cantidad')
plt.xticks(ticks=[0, 1], labels=['No', 'Sí'])
plt.show()
```

```python
# Histograma de la columna 'fecha'
plt.figure(figsize=(10, 6))
sns.histplot(data=df_reviews, x='fecha', bins=30, kde=True)
plt.title('Distribución de la fecha')
plt.xlabel('Fecha')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Gráfico de barras de la variable categórica 'reviews_recommend'
plt.figure(figsize=(8, 6))
sns.countplot(data=df_reviews, x='reviews_recommend')
plt.title('Distribución de reviews_recommend')
plt.xlabel('reviews_recommend')
plt.ylabel('Cantidad')
plt.show()
```

```python
# Gráfico de barras de la variable categórica 'reviews_posted'
plt.figure(figsize=(12, 6))
sns.countplot(data=df_reviews, x='reviews_posted')
plt.title('Distribución de reviews_posted')
plt.ylabel('Cantidad')
plt.show()
```

```python
# Gráfico de barras de la variable categórica 'sentiment_analysis'
plt.figure(figsize=(8, 6))
sns.countplot(data=df_reviews, x='sentiment_analysis')
plt.title('Distribución de sentiment analysis')
plt.xlabel('sentiment_analysis')
plt.ylabel('Cantidad')
plt.show()
```

```python
# Crear subplots
fig, axs = plt.subplots(1, 2, figsize=(15, 6))

# Histograma de la distribución de años
n, bins, _ = axs[0].hist(df_reviews['fecha'], bins=15, color='purple', edgecolor='purple', align='mid')
axs[0].set_title('Reseñas por año')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('Cantidad de Reseñas')
axs[0].grid(axis='y', linestyle='--', alpha=0.7)

# Gráfico de líneas para mostrar tendencias temporales
trends = df_reviews['fecha'].value_counts().sort_index()
axs[1].plot(trends.index, trends.values, marker='o', linestyle='-', color='red')
axs[1].set_title('Tendencia User Reviews')
axs[1].set_xlabel('Año')
axs[1].set_ylabel('Cantidad de Reseñas')
axs[1].grid(axis='y', linestyle='--', alpha=0.7)

# Ajustar el diseño
plt.tight_layout()

# Mostrar los gráficos
plt.show()
```
