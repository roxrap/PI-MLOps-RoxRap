```python
import pandas as pd
import numpy as np
import os
import json
```

```python
# Se abre el archivo JSON y se lee lÃ­nea por lÃ­nea. Cada lÃ­nea se convierte de JSON a un diccionario de Python utilizando json.loads(), y el diccionario resultante se agrega a la lista filas_games.
filas_games = []
with open(r'C:\Users\ROXANA\Downloads\04-2024\lanzamiento\Roxana\Datasets\output_steam_games.json') as f:
    for line in f.readlines():
        data = json.loads(line)
        filas_games.append(data)

# Se utiliza la lista de diccionarios filas_games para crear un DataFrame de pandas llamado df_games. Cada diccionario en la lista representa una fila en el DataFrame.
df_games = pd.DataFrame(filas_games)
df_games
```

```python
df_games.info()
```

```python
# Eliminar filas con todos los valores nulos
df_games = df_games.dropna(how='all').reset_index(drop=True)
df_games.shape
```

```python
# Se filtran las filas duplicadas
duplicated_rows = df_games[df_games.duplicated(subset='id', keep=False)]
duplicated_rows
```

```python
# se eliminan los index duplicados
index_a_eliminar = [74, 13894, 14573, 30961]
df_games = df_games.drop(index_a_eliminar)
```

```python
#transformar la columna de fecha
df_games['release_date'].value_counts()
```

```python
# Crea columna nueva con el aÃ±o
df_games['release_date'] = pd.to_datetime(df_games['release_date'], format='%Y-%m-%d', errors='coerce')
```

```python
df_games['release_date']
```

```python
#relleno los valores Nan con 0 para poder pasar a int
df_games['release_year'] = df_games['release_date'].dt.year.fillna(0).astype(int) 
```

```python
df_games['release_year']
```

```python
df_games.drop('release_date', axis=1, inplace=True) #elimino la columna 'release_date' ya q es reemplazada por 'release_year'
```

```python
data=df_games['price'].unique()  #busco los str en los prcios
data[np.array([isinstance(x, str) for x in data])]
```

```python
# Reemplazar los valores STR' con 0
df_games['price'] = df_games['price'].replace(['Free To Play', 'Free to Play', 'Free', 'Free Demo',
       'Play for Free!', 'Install Now', 'Play WARMACHINE: Tactics Demo',
       'Free Mod', 'Install Theme', 'Third-party', 'Play Now',
       'Free HITMANâ„¢ Holiday Pack', 'Play the Demo',
       'Starting at $499.00', 'Starting at $449.00', 'Free to Try',
       'Free Movie', 'Free to Use'], float('0'))

# Convertir la columna a tipo float
df_games['price'] = df_games['price'].astype(float)
```

```python
df_games['price']
```

```python
# Se rellenan los nulos
df_relleno = df_games[['publisher', 'app_name', 'title', 'developer']].fillna('Sin datos')
```

```python
# Se borran las columnas originales y se concatenan las rellenas con todo el dataframe
df_games = pd.concat([df_games.drop(['publisher', 'app_name', 'title', 'developer'], axis=1), df_relleno], axis=1)
df_games.head()
```

Este cÃ³digo se encarga de descomponer las listas o conjuntos anidados en la columna 'genres' en filas individuales mediante explode, elimina las filas que contienen valores nulos en 'genres' y muestra las primeras filas del DataFrame resultante.

```python
df_games = df_games.explode('genres')
df_games = df_games.dropna(subset=['genres'])
df_games.head()
```

```python
#elimino columnas q no voy a usar
df_games = df_games.drop(['tags', 'specs', 'url', 'reviews_url'], axis=1)
df_games.columns
```

```python
# Verificar si hay valores nulos en las columnas especificadas
columnas_verificar = ['genres', 'price', 'early_access', 'id', 'release_year', 'publisher', 'app_name', 'title', 'developer']
nulos = df_games[columnas_verificar].isnull().any()
nulos
```
