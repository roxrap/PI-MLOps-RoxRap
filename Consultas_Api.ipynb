```python
import pandas as pd
import os
import ast
```

```python
df_games = pd.read_csv('Datasets\games_limpios.csv')
df_reviews = pd.read_csv('Datasets\\reviews_limpios.csv')
df_items = pd.read_csv('Datasets\items_limpios.csv')
```

```python
df_items.head()
```

```python
df_games
```

```python
# Renombrar la columna 
df_games.rename(columns={'id': 'item_id'}, inplace=True)
df_games.rename(columns={'release_year': 'year'}, inplace=True)
```

```python
df_games['genres']
```

```python
df_merged = pd.merge(df_items, df_games[['item_id', 'year', 'genres']], on='item_id', how='inner')
```

```python
df_merged
```

```python
#Agrupamos por genero y año de lanzamiento
df_merged = df_merged.groupby(['year', 'genres'])['hours_game'].sum().reset_index()

```

```python
print(df_merged['genres'].unique())
```

```python
import html

# Decodificar entidades HTML en la columna 'genres'
df_merged['genres'] = df_merged['genres'].apply(html.unescape)
```

```python
print(df_merged['genres'].unique())
```

```python
# Reemplazar valores nulos con una cadena vacía
df_merged['genres'] = df_merged['genres'].fillna('')

# Eliminar filas con valores nulos en la columna 'genres'
df_merged.dropna(subset=['genres'], inplace=True)

# Encontrar el género con más horas jugadas por año 
max_hours_genre = df_merged.loc[df_merged.groupby(['year', 'genres'])['hours_game'].idxmax()]

# Ordenar el DataFrame resultante
max_hours_genre = max_hours_genre.sort_values(by=['genres', 'hours_game', 'year'], ascending=[True, False, True])

# Guardar el resultado en un archivo CSV
max_hours_genre.to_csv('PlayTimeGenre.csv', index=False)
```

```python
df_games
```

```python
merged_data=pd.merge(df_items,df_games,left_on='item_id', right_on='item_id', how='right')

merged_data.drop(['item_id'], axis=1, inplace=True)
merged_data.dropna(inplace=True)

merged_data['hours_game'] = merged_data['hours_game'].astype(int)
merged_data['year'] = merged_data['year'].astype(int)

#Encontrar el jugador con más horas jugadas por género y año
max_hours_df = merged_data.loc[merged_data.groupby(['genres', 'year'])['hours_game'].idxmax()]

#Filtrar las columnas necesarias
max_hours_df = max_hours_df[['genres', 'year', 'user_id', 'hours_game']]

#Segunda agrupación para obtener las horas jugadas por año para el usuario seleccionado
max_hours = max_hours_df.groupby(['genres', 'year', 'user_id'])['hours_game'].sum().reset_index()

max_hours.to_csv('UserForGenre.csv', index=False)
```

```python
# Filtrar las filas que cumplen con las condiciones
result = df_merged.loc[(df_merged['reviews_recommend'] == True) & (df_merged['sentiment_analysis'].isin([1, 2])),
                           ['fecha', 'app_name']]

# Agrupar por 'fecha' y 'name' y contar las ocurrencias
grouped_result = result.groupby(['fecha', 'app_name']).size().reset_index(name='count')

# Ordenar por 'fecha' y 'count' en orden descendente
grouped_result = grouped_result.sort_values(by=['fecha', 'count'], ascending=[False, False])

# Obtener el top 3 por año
recomendados_x_año = grouped_result.groupby('fecha').head(3)

# Guardar el resultado en un archivo CSV
recomendados_x_año.to_csv('UsersRecommend.csv', index=False)
```

```python
# Filtrar las filas que cumplen con las condiciones
result = df_merged.loc[(df_merged['reviews_recommend'] == True) & (df_merged['sentiment_analysis'].isin([0])),
                           ['fecha', 'app_name']]

# Agrupar por 'fecha' y 'name' y contar las ocurrencias
grouped_result = result.groupby(['fecha', 'app_name']).size().reset_index(name='count')

# Ordenar por 'fecha' y 'count' en orden descendente
grouped_result = grouped_result.sort_values(by=['fecha', 'count'], ascending=[False, False])

# Obtener el top 3 por año
no_recomendados_x_año = grouped_result.groupby('fecha').head(3)

# Guardar el resultado en un archivo CSV
no_recomendados_x_año.to_csv('UsersNotRecommend.csv', index=False)
```

```python
# Renombrar la columna 
df_reviews.rename(columns={'reviews_item_id': 'id'}, inplace=True)
```

```python
merged_df = pd.merge(df_games, df_reviews, on='id', how='inner')
```

```python
merged_df.head()
```

```python
# Agrupar por desarrolladora y sentimiento, y contar las ocurrencias
result_df = merged_df.groupby(['release_year', 'sentiment_analysis']).size().reset_index(name='count')

# Pivoteando el DataFrame para tener sentimientos como columnas
result_df = result_df.pivot(index='release_year', columns='sentiment_analysis', values='count').reset_index()

# Renombrar las columnas según el formato deseado
result_df.columns = ['release_year', 'Negative', 'Neutral', 'Positive']

# Rellenar NaN con 0
result_df = result_df.fillna(0)

# Convertir las columnas a tipo int
result_df[['Negative', 'Neutral', 'Positive']] = result_df[['Negative', 'Neutral', 'Positive']].astype(int)

# Guardar el resultado en un archivo CSV
result_df.to_csv('sentiment_analysis.csv', index=False)
```
